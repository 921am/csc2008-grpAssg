```python
import os
import pandas as pd
from datetime import datetime
from pprint import pprint
from pymongo import MongoClient
import sys
import time

def import_content(filepath, collection_name, counter_id, sort_name):
    global totalCreationTime
    global totalSeedingTime
    global totalExecutionTime
    global totalCounterTime

    file_res = os.path.basename(filepath)
    
    try:
        data = pd.read_csv(file_res)
        df = data.to_dict('records')
        
        collectionStart = datetime.now()
        collection = db[collection_name]
        collection.drop()        
        collectionEnd = datetime.now()
        
        seedingStart = datetime.now()
        collection.insert_many(df, ordered=False)
        seedingEnd = datetime.now()

        print("\n\n========================================", collection_name, "========================================\n\n")

        for i in collection.find():
            pprint(i)
            print('\n')

        counterStartTime = datetime.now()
        for i in collection.find().sort(sort_name, -1).limit(1):
            result = i[sort_name]
            
        col_counter.insert_one({'name': counter_id, 'seq': result})
        counterEndTime = datetime.now()
                
    except:
        col_counter.insert_one({'name': counter_id, 'seq': 0})
        print('No ', file_res, '\n')
    
    
    creationTimeDiff = collectionEnd - collectionStart
    seedingTimeDiff = seedingEnd - seedingStart
    totalTimeDiff = seedingTimeDiff + creationTimeDiff
    counterTimeDiff = counterEndTime - counterStartTime
    
    executionCreationTime = creationTimeDiff.total_seconds() * 1000
    executionSeedingTime = seedingTimeDiff.total_seconds() * 1000
    executionQueryTime = totalTimeDiff.total_seconds() * 1000
    counterTime = counterTimeDiff.total_seconds() * 1000
    
    totalCreationTime = executionCreationTime + totalCreationTime
    totalSeedingTime = executionSeedingTime + totalSeedingTime
    totalExecutionTime = executionQueryTime + totalExecutionTime
    totalCounterTime = counterTime + totalCounterTime
    
    print('Creation Time: ', executionCreationTime, 'ms')
    print('Seeding Time: ', executionSeedingTime, 'ms')
    print('Query Time: ', executionQueryTime, 'ms')
    print('Counter Time: ', counterTime, 'ms\n')
    
    print('Total Creation Time: ', totalCreationTime, 'ms')
    print('Total Seeding Time: ', totalSeedingTime, 'ms')
    print('Total Query Time: ', totalExecutionTime, 'ms')
    print('Total Counter Time: ', totalExecutionTime, 'ms\n')
    
    print('Seq count: ', result)
    print('Counter sequence')
    for i in col_counter.find():
        print(i)

        
if __name__ == "__main__":
    
    scriptStartTime = datetime.now()
    
    client = MongoClient(host='localhost', port=27017)
    db = client['CRUD']
    col_counter = db['Counter']
    col_counter.drop()
    
    totalCreationTime = 0
    totalSeedingTime = 0
    totalExecutionTime = 0
    totalCounterTime = 0
    
    collection_name = 'Inmates'
    sort_name = 'InmateID'
    counter_id = 'inmateID'
    filepath = '/inmates.csv'
    import_content(filepath, collection_name, counter_id, sort_name)
```
